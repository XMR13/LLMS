### Contoh penerapan mekanisme Attention
Akan ditampilkan 3 penerapan mekanisme atensi yang setiap versinya merupakan pengemabangan dari versi sebelumnya 
yang lebih lengkap
1. Self attention dengan weights manual
2. Self attention dengan trainanble weights dan linear layer
3. Causal attention with mask self attention dan dropout layer
4. Multihead attentioan 

```python
import torch.nn as nn
```

#### 1. Self Attention with trainable weights

```python
class Attention_Mechanism_V2(nn.Module):
    def __init__(self, d_in, d_out, qkv_bias=False):
        super().__init__()
        """
        Impelemntasi self attention yang telah teroptimisasi
        menggunakan layer linear dibandingkan dengan menggunakan 
        native parameter yang berasal dari nn.toch.Param
        """
        
        #initiate the weights 
        self.W_q = nn.Linear(in_features=d_in, out_features=d_out, bias=qkv_bias)
        self.W_k = nn.Linear(in_features=d_in, out_features=d_out, bias=qkv_bias)
        self.W_v = nn.Linear(in_features=d_in, out_features=d_out, bias=qkv_bias)


    #terapkan mekanisme atensinuya
    def forward(self, x):
        #create the matrix Q, K, V
        query = self.W_q(x)
        keys = self.W_k(x)
        values = self.W_v(x)    

        #lakukan perhitungan terhadap attention score
        attention_score = query @ keys.T
        attention_weights = torch.softmax(
            input=(attention_score / keys.shape[-1] ** 0.5),
            dim=-1
        )

        #menghitung konteks vektornya
        context_vector = attention_weights @ values
        return context_vector
```

##### 2. Masked Attention 
Adalah tipe attention yang digunakan pada bagian decoder di arstitektur transformer. Suatu input yang bisa mengakses token pada inputnya dan input sebelumnya, dan tidak bisa mengakses token setelahnya. Yang token setelah input tersebut akan *dimask* menjadi
0.

```python
class Masked_Attention(nn.Module):
    """Adalah pengembangan lanjut dari kelas attention sebelumnya,
        yang akan menambahkan fungsional
        1. Masked attention
        2. Dropout layer
        3. Dapat memrposes input dalam batch
    """

    def __init__(self, d_in, d_out, context_length,
                 dropout, kqv_bias = False):
        super().__init__()

        self.W_q = nn.Linear(in_features=d_in, out_features=d_out, bias=kqv_bias)
        self.W_k = nn.Linear(in_features=d_in, out_features=d_out, bias=kqv_bias)
        self.W_v = nn.Linear(in_features=d_in, out_features=d_out, bias=kqv_bias)
        self.dropout = nn.Dropout(dropout)
        self.register_buffer(
            'mask',
            torch.triu(torch.ones(context_length, context_length), diagonal=1)
        )

    def forward(self, x):

        #prepare the data with batch
        b, num_tokens, d_out = x.shape
        query = self.W_q(x)
        keys = self.W_k(x)
        values = self.W_v(x)

        #calulcate the attention scores
        att_scores = query @ keys.transpose(1,2)
        att_scores.masked_fill_(
            self.mask.bool()[:num_tokens, :num_tokens], -torch.inf
        )

        #calculate the attention weights
        att_weights = torch.softmax(att_scores/ keys.shape[-1] ** 0.5, dim=-1)

        #do the dropout
        att_weights = self.dropout(att_weights)

        #context vector
        context_vector = att_weights @ values

        return context_vector
```

#### 3. Multihead Attention:
jenis attention yang menggunakan beberapa head untuk menangkap hubungan dari berbagai macam context

```python
class Multihead_Attention_V2(nn.Module):
    def __init__(self, d_in, d_out, context_length, dropout,
                 num_heads, qkv_bias=False):
        super().__init__()

        #assert that
        assert(d_out % num_heads == 0), \
        'Pastikan jumlah heads adalah angka yang dapat dibagi dengan d_out'

        #menginisasi parameter
        self.d_in = d_in
        self.d_out = d_out
        self.num_heads = num_heads
        self.heads_dim = self.d_out // num_heads

        #the weights
        self.W_q = nn.Linear(in_features=d_in, out_features=d_out, bias = qkv_bias)
        self.W_k = nn.Linear(in_features=d_in, out_features=d_out, bias = qkv_bias)
        self.W_v = nn.Linear(in_features=d_in, out_features=d_out, bias = qkv_bias)
        self.out_proj = nn.Linear(d_out, d_out)
        self.dropout = nn.Dropout(dropout)

        #config
        self.register_buffer(
            'mask',
            torch.triu(torch.ones(context_length, context_length), diagonal = 1)
        )


    def forward(self, X):

        #1. pertama adalah tentukan dimensinya
        batch, num_tokens, d_in = X.shape

        #inisasikan QKV
        query = self.W_q(X)
        keys = self.W_k(X)
        values = self.W_v(X)


        #2,.ubah dimensinya sehingga dapat memiliki dimensi (batch, num_tokens, num_head, head_dim)
        query = query.view(batch, num_tokens, self.num_heads, self.heads_dim) 
        keys = keys.view(batch, num_tokens, self.num_heads, self.heads_dim) 
        values = values.view(batch, num_tokens, self.num_heads, self.heads_dim) 

        #kemudian transpose agar dapat diproses secara parallel untuk semua num_head nya (batch, num_heads, num_tokens, head_dim)
        query = query.transpose(1,2)
        keys = keys.transpose(1,2)
        vaues = values.transpose(1,2)


        #3. Hitung attention score nya
        att_score = query @ keys.transpose (2,3) #(num_tokens, head_dim) @ (head_dim, num_tokens)

        #4. Create the mask
        att_mask = self.mask.bool()[:num_token][:num_token] #buat mask dengan besar yang saama dengan token

        #5. Apply the mask
        att_score.masked_fill_(att_mask, -torch.inf)

        #6. Calculate the attention weights
        att_weights = torch.Softmax(
            att_score / keys.shape[-1] ** 0.5, dim=-1
        )

        #7. Do the dropout on the attetion weights
        att_weights = self.dropout(att_weights)

        #8. Calculate the context vector
        context_vector = (att_weights @ values).transpose(1,2) #ubah lagi menjadi (batch, num_tokens, num_head, head_dim)
        context_vector = context_vector.contiguous().view(
            batch, num_tokens, self.d_out
        )
        context_vector = self.out_proj(context_vector)

        return context_vector
```
